```python
df = pd.read_csv('alldata_1_for_kaggle.csv', encoding='latin-1')
df = df.drop(['Unnamed: 0'], axis=1)
df.columns = ['Class', 'Description']
df.tail()
```

```python
# Separate features and target
X = df.drop('Description', axis=1)
y = df['Class']

# Initialize StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1, test_size=2500, random_state=42)

# Perform the stratified split
for train_index, test_index in sss.split(X, y):
    stratified_sample = df.iloc[test_index]

print(stratified_sample)
```

```python
stratified_sample.to_csv('stratified_sample.csv', index=False)
```
